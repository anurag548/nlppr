```python
import spacy

#load the english language model
nlp = spacy.load('en_core_web_sm')

#process the text
doc = nlp("Elon Musk from USA is looking at buying Twitter.inc for $10Billion")

#iterate over the predicted entities
for ent in doc.ents:
  #print the entity text and its label
  print(ent.text, ent.label_)
```

```python
text = ('ISRO is a space research organization'
' situated in India headquatered at Bengaluru'
' the founder was Vikram Sarabhai.')
doc = nlp(text)
for ent in doc.ents:
  print(ent.text, ent.start_char, ent.end_char,
ent.label_, spacy.explain(ent.label))
```
